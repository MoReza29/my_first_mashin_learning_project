```python
import numpy as np
import pandas as pd
import seaborn as sns
import sklearn as sk
from sklearn import linear_model
import sklearn.metrics as sm
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
d=pd.read_csv('cardataset500.csv')
x=d.drop('output',axis=1)
y=d.output
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
model=linear_model.LinearRegression()
model.fit(x_train,y_train)

```

```python
a=float(input('what is the number of cylinders'))
while True:
    if 4<=a<=30:
        break
    if a>30:
        a=float(input('that number is too big for number of cylinders'))
    if a<4:
        a=float(input('that number is too small for number of cylinders'))
        
b=float(input('What is the engine size of your car?(Liter!)'))
while True:
    if 1<=b<=24:
        break
    if b>24:
        b=float(input('that number is too big for amount of engine size.What is the engine size of your car?(Liter!)'))
    if b<1:
        b=float(input('that number is too small for amount of engine size.What is the engine size of your car?(Liter!)'))
c=float(input('what is fuel consumption of your car'))
while True:
    if 1<=c<=100:
        break
    if c>100:
        c=float(input('that number is too big for amount of fuel consumption.What is the fuel consumption of your car?'))
    if c<1:
        c=float(input('that number is too small for amount of fuel consumption.What is the fuel consumption of your car?'))
machin=np.array([[a,b,c]])
co2=model.predict(machin)
print("the amount of produced co2 is",co2 )
```
